```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = 2*x
```

# Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='square')
ax.plot(x,x**3,label='cube')
ax.legend(loc = 0)

#0: dependent on matplotlib
#1: upper right corner
#2: upper left corner
#3: lower left corner
#4: lower right corner
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='square')
ax.plot(x,x**3,label='cube')
ax.legend(loc = (1.2, 0.5))
```

## Setting colors, linewidths, linetypes

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2,  'b.-' ,label='square')
ax.plot(x,x**3, 'go--',label='cube')
ax.legend(loc = 0)
```

# Colors with colors parameters

```python
fig, axes = plt.subplots()

axes.plot(x,x+1, color='#A46B6B')
axes.plot(x,x+3, color='#193300', alpha = 0.5)
```

```python

#linewidth
fig, axes = plt.subplots()

axes.plot(x,x+1, color='#A46B6B', lw= 0.6)
axes.plot(x,x+3, color='#193300', alpha = 0.5, lw = 10)
```

```python
#linestyles
fig, axes = plt.subplots()

axes.plot(x,x+1, color='#A46B6B', lw= 0.6, ls = '-')
axes.plot(x,x+3, color='#193300', alpha = 0.5, lw = 10, ls = '-.')
axes.plot(x,x+4, color='#193300', alpha = 0.5, lw = 10, ls = '--')
```

```python
#custom dash
fig, axes = plt.subplots()

lines = axes.plot(x,x+1, color='#A46B6B', lw= 0.6, ls = '-')
lines[0].set_dashes([5,3,10,10]) #(line length, space length)
```

```python
#custom dash
fig, axes = plt.subplots()
axes.plot(x,x+1, color='#A46B6B', lw= 0.6, marker= 's')
```

```python
#custom dash
fig, axes = plt.subplots()
axes.plot(x,x+1, color='#A46B6B', lw= 0.6, ls = '-' ,marker= 's', ms = 20,
         markerfacecolor = 'blue', markeredgewidth = 8, markeredgecolor= 'red')
```
