```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**3
```

```python
x = np.arange(0,10)
y = 2* x
```

# Creating a Figure

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) 
#Left, Bottom, Width, Height
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) 
#Left, Bottom, Width, Height
axes.plot(a,b)
plt.show()
```

# Add multiple axes to the figure

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])  #Large Figure
axes2 = fig.add_axes([0.15,0.35,0.5,0.5]) #Small Figure

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Big Figure')

axes2.plot(x,y)
axes2.set_title('Small Figure')

plt.show()
```

```python
fig =plt.figure(figsize=(12,4), dpi = 150)

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.15,0.35,0.5,0.5])
axes3 = fig.add_axes([1,1,0.1,0.1])

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Big Figure')

axes2.plot(x,y)
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Small Figure')

axes3.plot(a,b)
axes3.set_xlabel('X label')
axes3.set_ylabel('Y label')
axes3.set_title('Top Right Corner Figure')

plt.show()
```
