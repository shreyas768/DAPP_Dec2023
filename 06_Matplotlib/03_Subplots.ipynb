```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2* x
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,color='green', marker='D', linestyle='solid', linewidth=5)
```

## Adding rows and columns

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#empty canvas of 1 by 2 subplots
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
#empty canvas of 2 rows and 2 columns
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(a,b)
axes[1,0].plot(x,y)
#plt.tight_layout()   this automatically adjusts the positions of the axes on the figure canvas so that 
#there is no overlapping content
plt.show()
```

## Parameters on subplots()

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,7))

axes[0,0].plot(a,b)
axes[0,0].set_title('title of 0 0 ')
axes[0,0].set_xlabel('0 0 x label')

axes[1,1].plot(b,a)

axes[0,1].plot(x,y)
axes[0,1].set_ylabel('0 1 y label')

axes[1,0].plot(y,x)

fig.suptitle('this is figure level title', fontsize = 20 )
plt.show()
```

## Manual spacing on subplots()

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,7))

axes[0,0].plot(a,b)
axes[0,0].set_title('title of 0 0 ')
axes[0,0].set_xlabel('0 0 x label')

axes[1,1].plot(b,a)

axes[0,1].plot(x,y)
axes[0,1].set_ylabel('0 1 y label')

axes[1,0].plot(y,x)

fig.suptitle('this is figure level title', fontsize = 20 )


fig.subplots_adjust(left=0.2,
    bottom=None,
    right=0.5,
    top=None,
    wspace=0.9,
    hspace=0.5,
               )

fig.savefig('subplots.png',bbox_inches='tight')
plt.show()
```
