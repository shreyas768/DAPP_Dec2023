```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
x
```

```python
y
```

# Basic Matplotlib Commands

```python
plt.figure(figsize=(12,3))
plt.plot(x,y)
plt.xlabel('this is X axis')
plt.ylabel('this is Y axis')
plt.title('put title here')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('this is X axis')
plt.ylabel('this is Y axis')
plt.title('put title here')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

# Save a figure

```python
plt.plot(x,y)
plt.xlabel('this is X axis')
plt.ylabel('this is Y axis')
plt.title('put title here')
plt.xlim(0,6)
plt.ylim(0,12)
plt.savefig('xVSy.png')
```
