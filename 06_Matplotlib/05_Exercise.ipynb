```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
```

```python
c = 3*10**8  #speed of light
```

```python
E = m*c**2
```

```python
#plot  E v mass , xlim of 0 to 10
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E,'r',lw=6)
axes.set_title('Energy-mass diagram')
axes.set_xlabel('mass')
axes.set_ylabel('Energy')
axes.set_xlim(0,10)
```

```python
labels = ['1 mo','3 mo','6 mo', '1 yr','2 yr','3 yr', '5 yr', '7 yr', '10 yr', '20 yr', '30yr' ]
july16_2017 = [4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
#plotting both curves on same figure. add a legend

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(labels,july16_2017, label='16/07/2017')
axes.plot(labels,july16_2020, label='16/07/2020')
plt.legend()
plt.show()
```

```python
#plotting both curves on same figure. add a legend

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(labels,july16_2017, label='16/07/2017')
axes.plot(labels,july16_2020, label='16/07/2020')
plt.legend(loc=(1.02,0))
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,8))
axes[0].plot(labels,july16_2017,label='july16_2017',color='r',marker='o',ms=5)
axes[0].legend(loc=(1.2,0))

axes[1].plot(labels,july16_2020,label='july16_2020',color='g',marker='s',ms=5)
axes[1].legend(loc=(1.2,0))
plt.show()
```
