```python
df.drop(['address','phone','dish_liked','cuisines','reviews_list', 'menu_item',
         'listed_in(city)', 'menu_item' ],
        axis =1, inplace=True)
```

```python
def cleaned_rate(ip):
    if (ip == 'NEW' or ip == '-'):
        return np.nan
    else:
        ip = str(ip).split('/')
        ip = ip[0]
        return float(ip)

```

```python
df['rate'] = df['rate'].apply(cleaned_rate)
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['approx_cost(for two people)'].unique()
```

```python
def cleaned_cost(cst):
    cst = str(cst)
    if ',' in cst:
        cst = cst.replace(',','')
        return int(cst)
    else:
        return int(cst)
```

```python
df['approx_cost(for two people)'] = df[
    'approx_cost(for two people)'].apply(cleaned_cost)
```

```python
df['approx_cost(for two people)'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type_df = df['rest_type'].value_counts()
rest_type_df_below1000 = rest_type_df[rest_type_df<1000]
rest_type_df_below1000
```

```python
def cleaned_rest_type(ip):
    if (ip in rest_type_df_below1000):
        return 'Others'
    else: return ip
```

```python
df['rest_type'] = df['rest_type'].apply(cleaned_rest_type)
df['rest_type'].unique()
```

```python
df.sample(5)
```

```python
df['location'].value_counts()
```

```python
location_df = df['location'].value_counts()
location_df_below400 = location_df[location_df<400]
location_df_below400
```

```python
def cleaned_location(ip):
    if (ip in location_df_below400):
        return 'Others'
    else: return ip
```

```python
df['location'] = df['location'].apply(cleaned_location)
df['location'].value_counts()
```

### Cost of Restaurants

```python
df['approx_cost(for two people)'].value_counts()
```

```python
df.sample(2)
```

```python
rest_cost = df['approx_cost(for two people)'].groupby(df['name'],sort=True)
dict1 = {}

for i,j in df['name'].value_counts()[:10].to_dict().items():
    dict1[i] = round(rest_cost.get_group(i).mean(),2)

cost_df = pd.DataFrame(list(dict1.items()),
                      columns = ['Restaurant Names','Avg_2_people_cost'])
cost_df
```

```python
sns.barplot(data=cost_df.sort_values(by=['Avg_2_people_cost'],ascending=False), x= 'Avg_2_people_cost', y = 'Restaurant Names')
plt.title('Average cost(2 people) - Top 10 Restaurants')
```

# Most Famous Restaurant chain in Bengaluru

```python
rest_famous = df['rate'].groupby(df['name'],sort=True)
dict2 = {}

for i,j in df['name'].value_counts()[:10].to_dict().items():
    dict2[i] = round(rest_famous.get_group(i).mean(),2)

famous_df = pd.DataFrame(list(dict2.items()),
                      columns = ['Restaurant Names','Avg_Rating'])
famous_df
```

```python
sns.barplot(data=famous_df.sort_values(by=['Avg_Rating'],ascending=False), x= 'Avg_Rating', y = 'Restaurant Names')
plt.title('Avg_Rating - Top 10 Restaurants')
```
