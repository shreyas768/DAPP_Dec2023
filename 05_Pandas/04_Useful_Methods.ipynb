## .apply() method

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
def last_four_payID(n):
    return str(n)[-4:]
```

```python
df['Payment ID'][0]
```

```python
last_four_payID('Sun2959')
```

```python
df['last_four_ID'] = df['Payment ID'].apply(last_four_payID)
```

```python
df
```

## apply with lambda

```python
df['total_bill'].apply(lambda bill : bill * 1.18)
```

## Apply for multiple columns

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(df['total_bill'][0],df['tip'][0])
```

```python
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']), axis = 1)
```
