## pivot() method

```python
licenses = df[['Company','Product','Licenses']]
licenses
```

# pivot_table() method

```python
df
```

```python
pd.pivot_table(df,index='Company', aggfunc='sum')[['Licenses','Sale Price']]
```

```python
pd.pivot_table(df,index='Company', aggfunc='sum', values = ['Licenses','Sale Price'])
```

```python
df.columns
```

```python
#pivot table with index level 1 account manager, index level 2 contact, values 'sale price', mean aggregation
pd.pivot_table(df,index=['Account Manager','Contact'],values = ['Sale Price'], aggfunc='mean')
```

```python
pd.pivot_table(df,index=['Account Manager','Contact','Product'],
               values = ['Sale Price','Licenses'], aggfunc='mean', fill_value=0)
```
