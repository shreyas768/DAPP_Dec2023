# KDE plot

```python
np.random.seed(50)
sample_age = np.random.randint(0,100,200)
```

```python
df_age = pd.DataFrame(sample_age, columns=['age'])
```

```python
df_age.head()
```

```python
sns.rugplot(data=df_age, x='age')
```

```python
sns.displot(data=df_age,x='age',bins=10, rug=True, kde=True)
```

```python
sns.kdeplot(data=df_age,x='age')
```

```python
sns.kdeplot(data=df_age,x='age', clip = [0,100])
```

```python
#bandwidth (analagous to bins in histogram)
sns.kdeplot(data=df_age,x='age',bw_adjust=0.1)
```

```python
sns.kdeplot(data=df_age,x='age',bw_adjust=0.7)
```
