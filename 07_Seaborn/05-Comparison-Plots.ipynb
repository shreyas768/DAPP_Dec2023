```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\shrey\Desktop\DAPP_Aug2023\DATA\StudentsPerformance.csv")
df.head()
```

# Joint Plot

```python
sns.jointplot(x='math score',y='writing score',data=df)
```

```python
sns.set_palette('rocket')
sns.jointplot(x='math score',y='writing score',data=df, kind='hex')
```

```python
sns.set_palette('Set2')
sns.jointplot(x='math score',y='writing score',data=df, kind='kde')
```
