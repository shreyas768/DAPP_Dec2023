```python
#box plot
plt.figure(figsize=(12,5))
sns.boxplot(x='parental level of education', y = 'math score', data=df)
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='parental level of education', y = 'math score', data=df, hue = 'gender')
```

```python
plt.figure(figsize=(10,15))
sns.boxplot(y='parental level of education', x = 'math score', data=df, hue = 'gender', orient= 'h')
```

```python
#Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='reading score',data=df)
```

```python
#Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='reading score',data=df, hue='gender')
```

```python
#Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='reading score',data=df, hue='gender',split=True)
```

# Swarm Plot

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='reading score',data=df)
```
