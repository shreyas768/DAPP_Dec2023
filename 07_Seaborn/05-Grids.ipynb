# Catplot()

```python
sns.catplot(kind='box',x='gender',y='math score',data=df)
```

# Pairgrid

```python
b = sns.PairGrid(df, hue='gender')
b = b.map_upper(sns.scatterplot)
b = b.map_diag(sns.kdeplot, lw = 2)
b = b.map_lower(sns.kdeplot, colors='red')
b = b.add_legend();
```
