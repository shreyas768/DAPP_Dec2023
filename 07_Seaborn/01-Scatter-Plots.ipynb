```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

```python
df = pd.read_csv(r"C:\Users\shrey\Desktop\DAPP_Aug2023\DATA\dm_office_sales.csv")
df.head()
```

```python
df.info()
```

## Scatterplot

```python
sns.scatterplot(data=df, x='salary', y = 'sales')
```

## Seaborn Parameters

```python
#hue
sns.scatterplot(x='salary',y='sales',data=df, hue = 'division')
```

```python
#hue
sns.scatterplot(x='salary',y='sales',data=df, hue = 'work experience', palette='cividis')
```

```python
#size
plt.figure(figsize=(12,9))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'work experience', palette='cividis', size = 'training level')
plt.legend(loc=4)
plt.show()
```

```python
#size
plt.figure(figsize=(12,9))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'work experience', palette='cividis', size = 200)
plt.legend(loc=4)
plt.show()
```

```python
#alpha
plt.figure(figsize=(9,5))
sns.scatterplot(x='salary',y='sales',data=df, size = 500, alpha= 0.4)
plt.legend(loc=4)
plt.show()
```
